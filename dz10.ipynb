```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df_car = pd.read_csv(r"https://raw.githubusercontent.com/meleshko-anastasiia/DZ/main/Car%20Price.csv")
```

```python
df_car
```

```python
#1. Постройте столбчатую диаграмму
```

```python
df_car[df_car['Fuel'].isin(["CNG", "Diesel", "Electric", "LPG", "Petrol"])]
sns.barplot(data = df_car, x = 'Fuel', y = 'Selling_Price')
plt.show()
```

```python
# Исходя из графика можно сделать вывод, что машины на Дизельном топливе самые дорогие из продаваемого перечня.
```

```python
# Постройте гистограмму с распределением числовой переменной
```

```python
sns.histplot(x = 'Year', data = df_car, hue = 'Transmission', kde = True)
```

```python
# Исходя из графика можно сделать вывод, что в начале в продаже были только автомобили с механической коробкой передачи, 
# но начиная с 2000-х годов в продаже начали внедряться машины с автоматической коробкой передач. 
# На протяжении всех лет, лидером продаж являются машины с маханической коробкой передач.
```

```python
# Постройте точечную диаграмму для двух числовых переменных
```

```python
sns.scatterplot(data = df_car, x = 'Year', y = 'Selling_Price')
```

```python
# С течением времени, стоимость автомобилей увеличивалась. Самая дорогая продажа авто была зафиксирована в 2016 году.
```

```python
# Постройте тепловую карту для двух категориальных переменных
```

```python
df6 = df_car[df_car['Fuel'].isin(["CNG", "Diesel", "Electric", "LPG", "Petrol"])]
df6 = df6.groupby(['Fuel', 'Transmission'])['Selling_Price'].sum().unstack()/10000
sns.heatmap(df6, annot = True, fmt = '.3f')
```

```python
# Самыми дорогими авто являются машины на дизельном топливе с механической коробкой передач. 
# Самая низкая стоимость из проданных авто зафиксирована на электрокарах.
# Автомобили на СНГ(LPG) и КПГ(CNG) находятся в одной тепловой области, но стоимость машин на топливе КПГ(CNG) немного выше. 
```
