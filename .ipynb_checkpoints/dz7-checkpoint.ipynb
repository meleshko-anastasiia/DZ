```python
# 5. Заполните пропуски
```

```python
na_index = df_car[df_car['Year'].isna()].index
```

```python
mean_Year = df_car['Year'].mean()
df_car['Year_filled_mean'] = df_car['Year'].fillna(mean_Year)
df_car.loc[na_index, 'Year_filled_mean']
```

```python
median_Year = df_car['Year'].median()
df_car['Year_filled_median'] = df_car['Year'].fillna(median_Year)
df_car.loc[na_index, 'Year_filled_median']
```

```python
size_na = df_car['Year'].isna().sum()
random_Year = df_car['Year'].dropna().sample(size_na).values
df_car.loc[df_car['Year'].isna(), 'Year_filled_rand'] = random_Year 
```

```python
df_car.loc[na_index, 'Year_filled_rand']
```

```python
# 6.Проверить, есть ли в датасете выбросы. Если есть – удалить
```

```python
df_car.describe()
```

```python
df_car['Selling_Price'].hist(bins = 100)
```

```python
q1 = df_car['Selling_Price'].quantile(0.25)
q3 = df_car['Selling_Price'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
print(f"Выбросами будут считаться все продажи, у которых цена находится за пределами интервала [{lower_bound}, {upper_bound}]")
```

```python
df_car[(df_car['Selling_Price']<=lower_bound)|(df_car['Selling_Price']>=upper_bound)]
```

```python
df_car[(df_car['Selling_Price']>=lower_bound)&(df_car['Selling_Price']<=upper_bound)]
```

```python
# 7.Нормализовать данные одним из рассмотренных способов
```

```python
df_car['Selling_Price'] = (df_car['Selling_Price'] - df_car['Selling_Price'].min()) / (df_car['Selling_Price'].max() - df_car['Selling_Price'].min()) 
```
