```python
data = db.read_text('1.txt').str.split()
words = data.flatten()
word_count = words.frequencies()
pd.DataFrame(word_count, columns=['Word', 'Frequency'])
```

```python
df1 = pd.DataFrame(word_count, columns=['Word', 'Frequency'])
#Сортировка по убыванию Frequency
df1.sort_values(by=['Frequency'], ascending=False)
```

```python
df2 = df1.sort_values(by=['Frequency'], ascending=False)
df2.head(12)
```

```python
# В таблице представлены самые часто встречающиеся слова
# 10 самых часто встречающихся: и, в, на, не, с, он, что, его, как, к
```

```python
# Построить гистограмму распределения
```

```python
df3 = df2.head(12)
df3
```

```python
fig = plt.figure()
plt.title('Frequency')
sns.histplot(df3, kde = True)
plt.show()
```

```python
df3[df3['Word'].isin(["и", "в", "на", "не", "с"])]
sns.barplot(data = df3, x = 'Word', y = 'Frequency')
plt.show()
```
