```python
#Напишите другую функцию и примените ее к сгруппированным данным с помощью.transform()
```

```python
def get_lower_boundary(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    return lower_bound

def get_upper_boundary(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    return upper_bound
```

```python
df_all['lower_limit_for_replacing_the_oil'] = df_all.groupby('Model')['KM_Driven'].transform(get_lower_boundary)
df_all['upper_limit_for_replacing_the_oil'] = df_all.groupby('Model')['KM_Driven'].transform(get_upper_boundary)
```
